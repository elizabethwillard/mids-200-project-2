```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#clean 2016Q4 data
house2016Q4_detail = pd.read_csv("2016Q4-house-disburse-detail.csv")
personnelcompensation16Q4 = house2016Q4_detail.loc[house2016Q4_detail['CATEGORY'] == 'PERSONNEL COMPENSATION']
personnelcompensationb16Q4 = personnelcompensation16Q4.loc[personnelcompensation16Q4['AMOUNT'] > 0]
payee2016Q4 = personnelcompensationb16Q4.loc[:, ['OFFICE','PAYEE']].drop_duplicates(subset=['PAYEE'])
payee2016Q4['Period'] = '2016 Q4'
staffcount16Q4 = personnelcompensationb16Q4['PURPOSE'].value_counts().rename_axis('Position').reset_index(name='count')
staffcomp16Q4 = personnelcompensationb16Q4.groupby(['PURPOSE'])['AMOUNT'].agg('sum').rename_axis('Position').reset_index(name='total compensation')
staffsalary2016Q4 = pd.merge(left=staffcount16Q4, right=staffcomp16Q4, on='Position', how='inner')
staffsalary2016Q4['avg salary'] = staffsalary2016Q4['total compensation'] / staffsalary2016Q4['count']
staffsalary2016Q4['Period'] = '2016 Q4' 
staffsalary2016Q4.sort_values(['avg salary'], ascending=False).head(10)
staffsalary2016Q4 = staffsalary2016Q4[~staffsalary2016Q4['Position'].str.contains('(OVERTIME)')]
staffsalary2016Q4 = staffsalary2016Q4[~staffsalary2016Q4['Position'].str.contains('(OTHER COMPENSATION)')]
staffsalary2016Q4 = staffsalary2016Q4[~staffsalary2016Q4['Position'].str.contains('(ASST)')]
staffsalary2016Q4.sort_values(['avg salary'], ascending=True).head(15)
#personnelcompensation16Q4.loc[personnelcompensation16Q4['AMOUNT'] < 0]
```

```python
#calculate retention rate 2016
allstaff2016 = pd.concat([payee2016Q1,payee2016Q2,payee2016Q3,payee2016Q4])
retention2016 = allstaff2016.Period.groupby(allstaff2016.PAYEE).value_counts().unstack().fillna(0)
retention2016[retention2016 > 1] = 1
retention2016['total'] = retention2016[list(retention2016.columns)].sum(axis=1)
retention2016['retention rate'] = retention2016['total'] * 100 / 4
retention2016['retention rate'].describe()
```

```python
#calculate retention rate 2017
allstaff2017 = pd.concat([payee2017Q1,payee2017Q2,payee2017Q3,payee2017Q4])

retention2017 = allstaff2017.Period.groupby(allstaff2017.PAYEE).value_counts().unstack().fillna(0)
retention2017[retention2017 > 1] = 1
retention2017['total'] = retention2017[list(retention2017.columns)].sum(axis=1)
retention2017['retention rate'] = retention2017['total'] * 100 / 4
retention2017['retention rate'].describe()
```

```python
#calculate retention rate 2016-2017
allstaff = pd.concat([payee2016Q1,payee2016Q2,payee2016Q3,payee2016Q4,payee2017Q1,payee2017Q2,payee2017Q3,payee2017Q4])
retention = allstaff.Period.groupby(allstaff.PAYEE).value_counts().unstack().fillna(0)
retention[retention > 1] = 1
retention['total'] = retention[list(retention.columns)].sum(axis=1)
retention['retention rate'] = retention['total'] * 100 / 8
retention['retention rate'].describe()
```
